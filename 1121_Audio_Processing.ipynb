# Audio Processing

```python
# Import packages
import matplotlib.pyplot as plt
import librosa
import numpy as np
import soundfile as sf
```

## 1. Graph Plot

```python
# Coordinates
x = [1, 2, 3, 4, 5]
y = [-1, -2, -3, -4, -5]

# Plot
plt.plot(x, y)
plt.title('Simple Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.scatter(x, y)
plt.title('Simple Scatterplot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## 2. Fast Fourier Transform

```python
audio_file = "high.wav"
audio, sr = librosa.load(audio_file)
librosa.display.waveshow(audio, sr=sr)
```

```python
spectrum = np.fft.fft(audio)
freqs = np.fft.fftfreq(len(spectrum), d=1/sr)

plt.plot(freqs[:len(freqs) // 2], np.abs(spectrum)[:len(spectrum) // 2])
plt.title("Original FFT Magnitude")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Remove higher frequencies
highest_freq = 1000
spectrum[np.abs(freqs) > highest_freq] = 0

plt.plot(freqs[:len(freqs) // 2], np.abs(spectrum)[:len(spectrum) // 2])
plt.title("FFT Magnitude After Cutoff")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# spectrum -> audio
filtered_audio = np.real(np.fft.ifft(spectrum))
librosa.display.waveshow(filtered_audio, sr=sr)

# Save new audio
filtered_audio = librosa.util.normalize(filtered_audio)
sf.write(f"filtered_high.wav", filtered_audio, sr)
```
